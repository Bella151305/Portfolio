```python
# Collecting files names
def get_all_files(folder_path):
    file_names = []
    for root, directories, files in os.walk(folder_path):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            file_names.append(file_path)
    
    return file_names


PATH = 'wikiart-master/wiki_art_saved/meta'
files = get_all_files(PATH)

print(f'We have {len(files)} files with artists data.')
```

```python
# Collecting meta attributes
df_columns = []

for name in files:
    with open(name.replace('\\', '/'), 'rb') as infile:
        artist_data = json.load(infile)
    
    for el in artist_data:
        for k in el.keys():
            if k not in df_columns:
                df_columns.append(k)

print(f'We have {len(df_columns)} attributes of artwork.')
```

```python
# Create data frame with meta data
wikiart_df = pd.DataFrame(columns = ['path'] + df_columns)

for name in files:
    with open(name.replace('\\', '/'), 'rb') as infile:
        artist_data = json.load(infile)
        
    for el in artist_data:
        data = {col: np.nan for col in df_columns}
        data.update({'path': name.replace('\\', '/')})
        for k, v in el.items():
            data.update({k: v})

        wikiart_df = wikiart_df.append(data, ignore_index=True)
        
        
wikiart_df.info()
```

```python
def normalize_image_5000(image_path):
    output_path='wikiart-master/wiki_art_saved/resized_images_5000/'
    
    image = Image.open(image_path)
    image_resized = image.resize((256, 256), Image.ANTIALIAS) # reducing these distortions and smoothing out the edges of the resized image
    # Convert the image to 24-bit color depth
    image_resized = image_resized.convert("RGB")
    
    width, height = image_resized.size
    
    # Save the normalized image
    output_path_total = str(output_path+'_'.join(image_path.split('/')[-3:]))
    image_resized.save(output_path_total)
    image.close()
    
    return width, height, output_path_total


def normalize_image_120(image_path):
    output_path='wikiart-master/wiki_art_saved/resized_images_120/'
    
    image = Image.open(image_path)
    image_resized = image.resize((256, 256), Image.ANTIALIAS)
    # Convert the image to 24-bit color depth
    image_resized = image_resized.convert("RGB")
    
    width, height = image_resized.size
    
    # Save the normalized image
    output_path_total = str(output_path+'_'.join(image_path.split('/')[-3:]))
    image_resized.save(output_path_total)
    image.close()
    
    return width, height, output_path_total
```

```python
wikiart_5000_df['image_new_size'] = wikiart_5000_df['image_path'].apply(normalize_image_5000)
wikiart_5000_df['image_new_width'] = wikiart_5000_df['image_new_size'].apply(lambda x: x[0])
wikiart_5000_df['image_new_height'] = wikiart_5000_df['image_new_size'].apply(lambda x: x[1])
wikiart_5000_df['image_new_path'] = wikiart_5000_df['image_new_size'].apply(lambda x: x[2])

wikiart_5000_df[['image_new_width', 'image_new_height']].describe()
```
